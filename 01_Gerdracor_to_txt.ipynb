```python
import pandas
from lxml import etree
import glob
```

```python
metadata = pandas.read_csv('gerdracor-metadata.csv')
```

```python
metadata
```

```python
for name in metadata['name']:
    print(name)
    tree = etree.parse(f'gerdracor/tei/{name}.xml')

    # Remove namespace
    for elem in tree.getiterator():
        # Skip comments and processing instructions,
        # because they do not have names
        if not (
            isinstance(elem, etree._Comment)
            or isinstance(elem, etree._ProcessingInstruction)
        ):
            # Remove a namespace URI in the element's name
            elem.tag = etree.QName(elem).localname

    # Remove unused namespace declarations
    etree.cleanup_namespaces(tree)

    text_content = tree.xpath('//body//l//text() | //body//p//text()')

    # Join the text content into a single string
    text_string = ' '.join(text_content)

    # Print the resulting text string
    with open(f'gerdracor_txt/{name}.txt', 'w') as f:
        f.write(text_string)
```
